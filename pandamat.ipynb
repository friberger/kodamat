```python
df= df.drop(columns=['Livsmedelsnummer', 'Energi (kcal)']).dropna(axis='columns')
df
```

```python
from scipy.spatial.distance import squareform, pdist

df_dist = pd.DataFrame(squareform(pdist(df.iloc[:, 1:])), columns=df.Livsmedelsnamn.unique(), index=df.Livsmedelsnamn.unique())
```

```python
s = df_dist['Kokosfett']
z = list(zip(s, s.index))
z.sort()
```

```python
for i in range(20):
    print(z[i])
```
